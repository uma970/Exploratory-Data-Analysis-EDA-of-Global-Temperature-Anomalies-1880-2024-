```python
df = pd.read_csv('/content/Global_Temp.csv', skiprows=1)
```

```python
df = df.apply(pd.to_numeric, errors='coerce')
```

```python
# Drop any rows with NaN values if necessary
df.dropna(inplace=True)
```

```python
# Drop any duplicate rows also
df = df.drop_duplicates()
```

```python
df.head()
```

```python
df.describe()
```

```python
# Set plot style
sns.set_style("whitegrid")
```

```python
# Descriptive statistics summary
df_stats = df.describe()
```

```python
# Plot: Global Temperature Anomalies Over Time
plt.figure(figsize=(12, 6))
plt.plot(df["Year"], df["J-D"], label="Annual Mean (J-D)", color='red', linewidth=2)
plt.xlabel("Year")
plt.ylabel("Temperature Anomaly (C)")
plt.title("Global Temperature Anomalies (1880-2024)")
plt.legend()
plt.show()
```

```python
plt.figure(figsize=(12, 8))
sns.heatmap(df.iloc[:, 1:].corr(), annot=True, cmap="magma", fmt=".2f")
plt.title("Correlation Heatmap of Monthly & Seasonal Temperature Anomalies (Updated)")
plt.show()
```

```python
# Boxplot: Monthly Temperature Anomalies to Detect Outliers
plt.figure(figsize=(12, 6))
sns.boxplot(data=df.iloc[:, 1:13], palette="plasma")
plt.xlabel("Month")
plt.ylabel("Temperature Anomaly (C)")
plt.title("Boxplot of Monthly Temperature Anomalies (1880-2024)")
plt.show()
```

```python
# Seasonal Temperature Anomalies Over Time
plt.figure(figsize=(12, 6))
plt.plot(df["Year"], df["DJF"], label="Winter (DJF)", color='pink', linewidth=1)
plt.plot(df["Year"], df["MAM"], label="Spring (MAM)", color='yellow', linewidth=1)
plt.plot(df["Year"], df["JJA"], label="Summer (JJA)", color='blue', linewidth=1)
plt.plot(df["Year"], df["SON"], label="Fall (SON)", color='green', linewidth=1)
plt.xlabel("Year")
plt.ylabel("Temperature Anomaly (C)")
plt.title("Seasonal Temperature Anomalies Over Time (1880-2024)")
plt.legend()
plt.show()
```

```python
# Moving Average Plot (10-Year)
df["10-Year MA"] = df["J-D"].rolling(window=10).mean()

plt.figure(figsize=(12, 6))
plt.plot(df["Year"], df["J-D"], label="Annual Mean (J-D)", color='lightgray', linewidth=1)
plt.plot(df["Year"], df["10-Year MA"], label="10-Year Moving Average", color='purple', linewidth=2)
plt.xlabel("Year")
plt.ylabel("Temperature Anomaly (C)")
plt.title("10-Year Moving Average of Global Temperature Anomalies")
plt.legend()
plt.show()
```

```python
# Moving Average Plot (5-Year)
df["5-Year MA"] = df["J-D"].rolling(window=10).mean()

plt.figure(figsize=(12, 6))
plt.plot(df["Year"], df["J-D"], label="Annual Mean (J-D)", color='lightgray', linewidth=1)
plt.plot(df["Year"], df["5-Year MA"], label="5-Year Moving Average", color='yellow', linewidth=2)
plt.xlabel("Year")
plt.ylabel("Temperature Anomaly (C)")
plt.title("5-Year Moving Average of Global Temperature Anomalies")
plt.legend()
plt.show()
```
